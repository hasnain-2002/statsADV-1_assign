```python
#ANS 9)
import seaborn as sns

def compute_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return (data - mean) / std

data = np.random.normal(50, 10, 100)
z_scores = compute_z_scores(data)

sns.histplot(z_scores, bins=20, kde=True)
plt.title("Z-scores of Sample Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```

```python
#ans 12)

import numpy as np
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    z = (x - mean) / std
    probability = norm.cdf(z)

    print(f"Data Point: {x}")
    print(f"Mean: {mean}, Standard Deviation: {std}")
     print(f"Z-score: {z:.4f}")
    print(f"Probability that a value is ‚â§ {x} (P(Z ‚â§ {z:.2f})): {probability:.4f}")
```

```python
#ans 13)
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate Z-statistic
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Two-tailed test: calculate p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-statistic = {z:.4f}")
    print(f"P-value = {p_value:.4f}")

    if p_value < alpha:
        print("Result: Reject the null hypothesis (statistically significant).")
    else:
        print("Result: Fail to reject the null hypothesis (not statistically significant).")


```

```python
#ans 14)
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation
    z_critical = norm.ppf((1 + confidence) / 2)  # Z-score for desired confidence level

    margin_of_error = z_critical * (std / np.sqrt(n))
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample Mean: {mean:.2f}")
    print(f"Confidence Interval ({int(confidence*100)}%): [{lower_bound:.2f}, {upper_bound:.2f}]")
    return lower_bound, upper_bound


```

```python
#ans 15)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0      # mean
sigma = 1   # standard deviation

# Generate x values (range of values)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Compute the PDF
pdf = norm.pdf(x, mu, sigma)

# Plot the PDF
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label=f'N({mu}, {sigma}¬≤)', color='blue')
plt.title('Probability Density Function of a Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#ans 16)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameter for the Poisson distribution (average rate Œª)
lambda_ = 5

# Generate x values (possible event counts)
x = np.arange(0, 15)

# Calculate the CDF for each value of x
cdf = poisson.cdf(x, lambda_)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf, where='post', label=f'Poisson CDF (Œª={lambda_})', color='green')
plt.title('Cumulative Distribution Function of a Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#ans 18)
import numpy as np
import matplotlib.pyplot as plt

# Generate two datasets
# Dataset 1: Standard deviation = 3
data1 = np.random.normal(loc=50, scale=3, size=500)

# Dataset 2: Standard deviation = 8
data2 = np.random.normal(loc=50, scale=8, size=500)

# Calculate the standard deviations
std_dev_data1 = np.std(data1)
std_dev_data2 = np.std(data2)

# Print the standard deviations
print(f"Standard Deviation of Dataset 1: {std_dev_data1:.2f}")
print(f"Standard Deviation of Dataset 2: {std_dev_data2:.2f}")

# Visualize the difference using histograms
plt.figure(figsize=(10, 6))

# Plot histogram for Dataset 1
plt.hist(data1, bins=30, alpha=0.5, label=f'Dataset 1 (œÉ={std_dev_data1:.2f})', color='blue')

# Plot histogram for Dataset 2
plt.hist(data2, bins=30, alpha=0.5, label=f'Dataset 2 (œÉ={std_dev_data2:.2f})', color='orange')

# Add title and labels
plt.title('Comparison of Standard Deviations of Two Datasets')
plt.xlabel('Data Values')
plt.ylabel('Frequency')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
